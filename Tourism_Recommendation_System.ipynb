```python
dataset  = pd.read_csv("/content/India Cities LatLng.csv")
df  = pd.DataFrame(dataset)
df
```

```python
df = df[["city","lat","lng"]]
df = df.rename(columns={"city":"City","lat":"Latitude","lng":"Longitude"})
df
```

```python
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    return only_ascii.decode('utf-8')
```

```python
#Removing from data_INDIA
df['City'] = df['City'].apply(remove_accents)
df
```

```python
kmeans = KMeans(100)
kmeans.fit(df[["Latitude","Longitude"]])

identified_clusters = kmeans.fit_predict(df[["Latitude","Longitude"]])
df["Loc_Cluster"] = list(identified_clusters)
df
```

```python
input_city = input("Enter city you visted: ")
cluster = df.loc[df["City"] == input_city, "Loc_Cluster"]
cluster = cluster.iloc[0]
cities = df.loc[df["Loc_Cluster"] == cluster, "City"]
recommended = {}
for i in range(len(cities)):
    if cities.iloc[i] == input_city or cities.iloc[i] in recommended:
        continue
    else:
        recommended[i] = cities.iloc[i]
recommended
```
